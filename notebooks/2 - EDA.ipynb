```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../data/raw/loan_data.csv")
dataset.head()
```

### Estad√≠sticas descriptivas

```python
dataset.describe().T
```

### Valores Nulos

```python
dataset.isnull().mean()
```

```python
col_numericas = dataset.select_dtypes(include = ['int64', 'float64'])
```

### Matriz de Correlaci√≥n

```python
plt.figure(figsize=(10, 8))
corr = col_numericas.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de correlaci√≥n")
plt.show()
```

### Gr√°ficas Variables Num√©ricas

```python
for col in col_numericas:
    plt.figure(figsize = (6, 4))
    sns.histplot(dataset[col], kde = True, bins=30)
    plt.title(f"Distribuci√≥n {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()
```

### Boxplots

```python
# Loan Amount y Person Gender
plt.figure(figsize=(12, 10))
sns.boxplot(data=dataset, x='person_gender', y='loan_amnt', palette="viridis")
plt.title("Distribuci√≥n de Loan Amount por G√©nero")
plt.xlabel("G√©nero")
plt.ylabel("Loan Amount")
plt.show()

# Ingresos y Nivel Educativo
plt.figure(figsize=(12, 10))
sns.boxplot(data=dataset, x='person_education', y='person_income', palette="viridis")
plt.title("Distribuci√≥n de Ingresos por Nivel Educativo")
plt.xlabel("Nivel Educativo")
plt.ylabel("Ingresos")
plt.show()

# Tipo de Propiedad y Edad
plt.figure(figsize=(12, 10))
sns.boxplot(data=dataset, x='person_home_ownership', y='person_age', palette="viridis")
plt.title("Distribuci√≥n de Tipos de Propiedad por Edad")
plt.xlabel("Tipo de Propiedad")
plt.ylabel("Edad")
plt.show()

# Monto de Cr√©dito y Prop√≥sito de Cr√©dito
plt.figure(figsize=(12, 10))
sns.boxplot(data=dataset, x='loan_intent', y='loan_amnt', palette="viridis")
plt.title("Distribuci√≥n de Monto de Cr√©dito por Prop√≥sito del Cr√©dito")
plt.xlabel("Prop√≥sito del Cr√©dito")
plt.ylabel("Monto del Cr√©dito")
plt.show()
```

### Gr√°fico de Barras Para Variables Categ√≥ricas

```python
col_categoricas = dataset.select_dtypes(include = ['object']).columns
col_categoricas
```

```python
for col in col_categoricas:
    plt.figure(figsize = (6, 4))
    sns.countplot(data = dataset, x = col, palette = 'viridis')
    plt.title(f"Frecuencia {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()
```
